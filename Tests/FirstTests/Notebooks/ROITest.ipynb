```python
import math
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt
# Load your image
```

```python
image = cv2.imread('back_Airpods.jpg')

imgrgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


plt.imshow(imgrgb)
plt.axis('off') 
plt.title('original')
plt.show()

```

```python
height, width = image.shape[:2]

# Calculate the center coordinates
center_x = width // 2
center_y = height // 2

# Example radius
radius = 1100



## Create a circular ROI in the mask
roi = np.zeros(image.shape[:2], np.uint8)


roi = cv2.circle(roi, (center_x, center_y), radius, 255, cv2.FILLED)


plt.imshow(roi)
plt.axis('off') 
plt.title('original')
plt.show()
```

```python
mask = np.ones_like(image) * 255
plt.imshow(mask)
plt.axis('off') 
plt.title('original')
plt.show()
```

```python
result = cv2.bitwise_and(image, image, mask=roi)
plt.imshow(result)
plt.axis('off') 
plt.title('original')
plt.show()
```
