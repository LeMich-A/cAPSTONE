```python
import math
import cv2
import numpy as np
import os
from matplotlib import pyplot as plt

```

```python

# Load the images
image1 = cv2.imread('finger1.jpg')
image2 = cv2.imread('finger2.jpg')


weight1 = 0.7
weight2 = 0.3

# Compute the blended image
blendedImage = cv2.addWeighted(image1, weight1, image2, weight2, 0)

# Clip pixel values (optional)
blendedImage = np.clip(blendedImage, 0, 255).astype(np.uint8)








# Display matched keypoints
plt.imshow(blendedImage)
plt.axis('off')
plt.title("merged image")
```

```python
# Load the images
image1 = cv2.imread('finger1.jpg')
image2 = cv2.imread('finger2.jpg')


image1rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB) 
image2rgb = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB) 
plt.subplot(1, 2, 1)
plt.imshow(image1rgb)
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(image2rgb)
plt.axis('off')
```

```python
# Ensure both images have the same dimensions
image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Set the weight for each image (you can adjust these values according to your preference)
alpha = 0.5  # Weight for the first image
beta = 0.5   # Weight for the second image

# Merge the images using addWeighted function
merged_image = cv2.addWeighted(image1, alpha, image2, beta, 0)
# merged_image = cv2.bitwise_and(image1, image2)
mergedrgb = cv2.cvtColor(merged_image, cv2.COLOR_BGR2RGB) 

# Display the merged image
plt.imshow(mergedrgb)
plt.axis('off')
plt.title("merged image")
```
